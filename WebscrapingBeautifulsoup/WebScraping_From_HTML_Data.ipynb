**The code will be modified to get the executable links**

- An empty set will be used to store the link

- Set has been used instead of list beacuse it will not allow duplicate links

```python
anchor = soup.find_all ('a')
all_links = set ()
```

```python
for link in anchor:
  if (link.get ('href') != '#'): # There are certain links that contain " # ", that are not required
    link1 = "http://www.codewithharry.com" +link.get ('href')
    all_links.add (link1)
    print (link1)
```

**Extracting the elements of specific div**

```python
s = soup.find (id = "nav-content")
#print (s.children)
print (s.content)
```

- .content ---- A tag's children return as a list
- using content the elements will be saved in memory
- it is slow due to memory usage


---------------------------------------------------------

- .children --- A tags children return as a generator/iterable
- using children elements will not save in meory but can be accessible anytime
- it is fast

```python
s = soup.find (id = "nav-content")
for elem in s:
  print (elem)
```

**To get the strings in any div**

```python
for item in s.strings:
  print (item)
```

**To print parent of any tag**

```python
s = soup.find (id = "nav-content")
print (s.parent)
```

If we use ".parents" instead of ".parent" we will get an iterable object

```python
s = soup.find (id = "nav-content")
for d in s.parents:
  print (d)
```

**To get name of parents**

```python
print (d.name)
```
